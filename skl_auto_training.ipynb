```python
import warnings

from configs_algorithms import ALGORITHMS_CONFIG_SKL_AUTO
from configs_data import DATASETS_CONFIG
from skl_auto_wrapper import SKLAutoWrapper
```

```python
# ignore unproblematic parameter warnings
warnings.filterwarnings("ignore")
```

```python
# show order of used datasets
for data_config in DATASETS_CONFIG:
    print(data_config["dataset"])
```

```python
# show order of used algorithms
for algorithm_config in ALGORITHMS_CONFIG_SKL_AUTO.values():
    print(algorithm_config.__name__)
```

```python
# execute sklearn automl over all algorithms for each dataset
sklauto_all: list[SKLAutoWrapper] = []
for data_config in DATASETS_CONFIG:
    sklauto_model = SKLAutoWrapper(data_config)
    sklauto_model.find_best_algorithm_poly()
    sklauto_all.append(sklauto_model)
```

```python
# print summarized execution results of sklearn automl for each dataset
[print(sklauto_run.get_search_summary()) for sklauto_run in sklauto_all]
```

```python
# print final best model of sklearn automl for each dataset
[display(sklauto_run.get_best_model()) for sklauto_run in sklauto_all]
```
